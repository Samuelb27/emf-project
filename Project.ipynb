```python
import pandas as pd
import numpy as np
```

```python
path = '/Users/ahmadroukain/Desktop/Cours/S2/Empirical methods/Project 1/Data_Project1.xlsx'
df = pd.read_excel(path, sheet_name='Entertainment', index_col=0)
df.drop(index='Code', inplace=True)
df.dropna(inplace=True)
```

```python
daily_simple_returns = df.pct_change().dropna()
daily_log_returns = pd.DataFrame()
for i in df.columns:
    daily_log_returns[i] = df[i].apply(lambda x : np.log(x)) - df[i].apply(lambda x : np.log(x)).shift(1)
daily_log_returns.dropna(inplace=True)
# for i in df.columns:
#     np.log(df[i])
# df['ACTIVISION BLIZZARD DEAD - DELIST.16/10/23 (#T) - ACTIVISION BLIZZARD DEAD - DELIST.16/10/23 (#T)'].apply(lambda x : np.log(x))
# df['ACTIVISION BLIZZARD DEAD - DELIST.16/10/23 (#T) - ACTIVISION BLIZZARD DEAD - DELIST.16/10/23 (#T)'].fillna(5, inplace=True)
# np.log(df['ACTIVISION BLIZZARD DEAD - DELIST.16/10/23 (#T) - ACTIVISION BLIZZARD DEAD - DELIST.16/10/23 (#T)'])
```

```python
weekly_df = df.iloc[3::5]
weekly_simple_returns = weekly_df.pct_change().dropna()
weekly_log_returns = pd.DataFrame()
for i in df.columns:
    weekly_log_returns[i] = weekly_df[i].apply(lambda x : np.log(x)) - weekly_df[i].apply(lambda x : np.log(x)).shift(1)
weekly_log_returns.dropna(inplace=True)
```

```python
ss = pd.DataFrame()
x = daily_simple_returns
ss['Mean'] = x.mean()
ss['Variance'] = x.var()
ss['Skewness'] = x.skew()
ss['Kurtosis'] = x.kurtosis()
ss['Min'] = x.min()
ss['Max'] = x.max()
ss
```
